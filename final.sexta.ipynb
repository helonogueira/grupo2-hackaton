## Cervejas separadas por semana

```python
import pandas as pd
from pyxlsb import open_workbook

FILE_PATH = r"C:\Users\ferna\Downloads\AMBEV\separado.por.cerveja.xlsb"

def get_bloco_dict(header, start, next_start):
    end = next_start if next_start else len(header)
    result = {}
    for i in range(start, end):
        val = str(header[i]).strip()
        if val not in ["", "None", "nan"]:
            result[val] = i
    return result

def ler_aba(sheet_name):
    rows = []
    with open_workbook(FILE_PATH) as wb:
        with wb.get_sheet(sheet_name) as sheet:
            for row in sheet.rows():
                rows.append([c.v for c in row])

    raw = pd.DataFrame(rows)

    header_row_idx = None
    for i, row in raw.iterrows():
        if row.astype(str).str.strip().str.lower().eq("demanda").any():
            header_row_idx = i
            break

    header = raw.iloc[header_row_idx].tolist()
    data   = raw.iloc[header_row_idx + 1:].reset_index(drop=True)

    semana_starts = [i for i, h in enumerate(header) if str(h).strip().lower() == "demanda"]
    COL_REGIAO    = next(i for i, h in enumerate(header) if str(h).strip() not in ["", "None", "nan"])

    records = []
    for idx_semana, col_start in enumerate(semana_starts):
        next_start = semana_starts[idx_semana + 1] if idx_semana + 1 < len(semana_starts) else None
        bloco = get_bloco_dict(header, col_start, next_start)

        def get_col(row, nomes):
            for n in nomes:
                if n in bloco:
                    return pd.to_numeric(row.iloc[bloco[n]], errors="coerce")
            return None

        for _, row in data.iterrows():
            regiao = row.iloc[COL_REGIAO]
            if pd.isna(regiao) or str(regiao).strip() in ["", "None", "TOTAL"]:
                continue

            records.append({
                "Cerveja":       sheet_name,
                "Regiao":        str(regiao).strip(),
                "Semana":        f"W{idx_semana}",
                "Demanda":       get_col(row, ["Demanda"]),
                "WSNP":          get_col(row, ["WSNP", "Malha", "0.0"]),
                "EI_Semana":     get_col(row, ["EI Semana"]),
                "Suf_in":        get_col(row, ["Suf.ini(d)"]),
                "Transi_Intern": get_col(row, ["Transf. Interna"]),
                "Transi_Cabo":   get_col(row, ["Transf. Ext (Cabo)"]),
                "Transi_Rodo":   get_col(row, ["Transf. Ext (Rodo)"]),
                "Transito":      get_col(row, ["Trânsito"]),
                "EF_Semana":     get_col(row, ["EF Semana"]),
                "Suf_f":         get_col(row, ["Suf. f (d)"]),
            })

    return pd.DataFrame(records)

# ── Ler todas as abas ─────────────────────────────────────────────────────────
abas = ["patagonia", "goose", "malzbier ", "colorado "]
df_all = pd.concat([ler_aba(a) for a in abas], ignore_index=True)

# ── Exibir por cerveja e semana ───────────────────────────────────────────────
for cerveja, df_cerv in df_all.groupby("Cerveja"):
    print(f"\n{'#'*60}")
    print(f"  {cerveja.upper()}")
    print(f"{'#'*60}")
    for semana, grupo in df_cerv.groupby("Semana"):
        print(f"\n{'='*60}  {semana}  {'='*60}")
        display(grupo.reset_index(drop=True))
ei_suf_w0 = df_all[df_all["Semana"] == "W0"][["Cerveja", "Regiao", "EI_Semana", "Suf_in"]]

```

## Redistribuicao da producao no neno

```python
producao = {
    "patagonia": {"CE": {"W0": 12240, "W1": 1800,  "W2": 5040,  "W3": 12600},
                  "PE": {"W0": 0,     "W1": 0,      "W2": 0,     "W3": 0}},
    "malzbier ": {"CE": {"W0": 0,     "W1": 9000,   "W2": 7560,  "W3": 0},
                  "PE": {"W0": 16200, "W1": 0,      "W2": 12960, "W3": 0}},
    "colorado ": {"CE": {"W0": 0,     "W1": 0,      "W2": 0,     "W3": 0},
                  "PE": {"W0": 5400,  "W1": 0,      "W2": 10800, "W3": 0}},
    "goose":     {"CE": {"W0": 0,     "W1": 0,      "W2": 0,     "W3": 0},
                  "PE": {"W0": 5400,  "W1": 14400,  "W2": 0,     "W3": 12600}},
}


def analisar_cerveja(cerveja, df_all, producao):
    df_cerv = df_all[df_all["Cerveja"] == cerveja].copy()
    ei_w0   = df_cerv[df_cerv["Semana"] == "W0"].set_index("Regiao")["EI_Semana"].to_dict()
    semanas = ["W0", "W1", "W2", "W3"]
    regioes = sorted(df_cerv["Regiao"].unique().tolist())

    dem_dict  = {}
    cabo_dict = {}
    for sem in semanas:
        sub = df_cerv[df_cerv["Semana"] == sem].set_index("Regiao")
        dem_dict[sem]  = sub["Demanda"].to_dict()
        cabo_dict[sem] = sub["Transi_Cabo"].fillna(0).to_dict()

    estoque    = {r: float(ei_w0.get(r, 0) or 0) for r in regioes}
    resultados = []

    for i, sem in enumerate(semanas):
        prod_disp = producao[cerveja]["CE"].get(sem, 0) + producao[cerveja]["PE"].get(sem, 0)
        prox_sem  = semanas[i + 1] if i + 1 < len(semanas) else None

        nec = {}
        for r in regioes:
            dem      = dem_dict[sem].get(r, 0)
            ei       = estoque[r]
            t_cabo   = cabo_dict[sem].get(r, 0)
            dem_prox = dem_dict[prox_sem].get(r, 0) if prox_sem else dem
            ef_alvo  = 12 * (dem_prox / 6)
            ef_base  = ei + t_cabo - dem
            envio_nec = max(0, ef_alvo - ef_base)
            nec[r] = {"dem": dem, "ei": ei, "t_cabo": t_cabo,
                      "ef_alvo": ef_alvo, "envio_nec": envio_nec, "dem_prox": dem_prox}

        total_nec = sum(v["envio_nec"] for v in nec.values())

        for r in regioes:
            dem       = nec[r]["dem"]
            ei        = nec[r]["ei"]
            t_cabo    = nec[r]["t_cabo"]
            ef_alvo   = nec[r]["ef_alvo"]
            envio_nec = nec[r]["envio_nec"]
            dem_prox  = nec[r]["dem_prox"]

            envio_prod = prod_disp * (envio_nec / total_nec) if total_nec > 0 else 0
            vem_sp     = max(0, envio_nec - envio_prod)

            if sem in ["W0", "W1", "W2"]:
                modal         = "Rodoviário"
                vol_enviar_sp = vem_sp / 0.95
            else:
                modal         = "Cabotagem"
                vol_enviar_sp = vem_sp

            ef    = ei + envio_prod + vem_sp + t_cabo - dem
            suf_f = ef / (dem_prox / 6) if dem_prox > 0 else 0

            resultados.append({
                "Cerveja":       cerveja,
                "Semana":        sem,
                "Regiao":        r,
                "Demanda":       round(dem, 2),
                "Dem_Proxima":   round(dem_prox, 2),
                "EI":            round(ei, 2),
                "Transi_Cabo":   round(t_cabo, 2),
                "Prod_NE":       round(envio_prod, 2),
                "Vem_SP":        round(vem_sp, 2),
                "Vol_Enviar_SP": round(vol_enviar_sp, 2),
                "Modal_SP":      modal if vem_sp > 0 else "-",
                "EF":            round(ef, 2),
                "Suf_f":         round(suf_f, 1),
            })
            estoque[r] = ef

    return pd.DataFrame(resultados)

cervejas = ["patagonia", "malzbier ", "colorado ", "goose"]
df_final = pd.concat([analisar_cerveja(c, df_all, producao) for c in cervejas], ignore_index=True)

for cerveja, df_c in df_final.groupby("Cerveja"):
    print(f"\n{'#'*60}  {cerveja.upper()}  {'#'*60}")
    for sem, grupo in df_c.groupby("Semana"):
        print(f"\n{'='*50}  {sem}  {'='*50}")
        display(grupo.reset_index(drop=True))

print("\n=== RESUMO — O QUE VEM DE SP ===")
sp = df_final[df_final["Vem_SP"] > 0][
    ["Cerveja","Semana","Regiao","Demanda","Transi_Cabo","Prod_NE","Vem_SP","Vol_Enviar_SP","Modal_SP","Suf_f"]]
display(sp if len(sp) > 0 else pd.DataFrame([{"Resultado": "Nenhuma transferência necessária!"}]))

print("\n=== PRODUÇÃO TOTAL NE POR CERVEJA/SEMANA ===")
display(df_final.groupby(["Cerveja","Semana"])["Prod_NE"].sum().reset_index())
```

```python
import json

regioes  = ["Mapapi", "NE Norte", "NE Sul", "NO Araguaia", "NO Centro"]
cervejas = ["patagonia", "malzbier ", "colorado ", "goose"]
semanas  = ["W0", "W1", "W2", "W3"]

def get_suf_depois(df, cerv, sem, reg_list):
    sub = df[(df["Cerveja"]==cerv) & (df["Semana"]==sem)].set_index("Regiao")
    return [round(float(sub.loc[r,"Suf_f"]),1) if r in sub.index else 0 for r in reg_list]

def get_suf_antes(df, cerv, sem, reg_list):
    sub = df[(df["Cerveja"]==cerv) & (df["Semana"]==sem)].set_index("Regiao")
    vals = []
    for r in reg_list:
        if r not in sub.index:
            vals.append(0); continue
        row      = sub.loc[r]
        ef_orig  = pd.to_numeric(row.get("EF_Semana", 0), errors="coerce") or 0
        dem_prox = pd.to_numeric(row.get("Dem_Proxima", row.get("Demanda", 1)), errors="coerce") or 1
        vals.append(round(float(ef_orig) / (float(dem_prox) / 6), 1))
    return vals

antes  = {}
depois = {}
for c in cervejas:
    key = c.strip()
    antes[key]  = {}
    depois[key] = {}
    for s in semanas:
        antes[key][s]  = get_suf_antes(df_all,   c, s, regioes)
        depois[key][s] = get_suf_depois(df_final, c, s, regioes)

data_json = json.dumps({
    "antes": antes, "depois": depois,
    "regioes": regioes, "semanas": semanas,
    "cervejas": [c.strip() for c in cervejas]
})
print("JSON gerado com sucesso")
print(data_json[:200])
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np

regioes  = ["Mapapi", "NE Norte", "NE Sul", "NO Araguaia", "NO Centro"]
cervejas = ["patagonia", "malzbier ", "colorado ", "goose"]
semanas  = ["W0", "W1", "W2", "W3"]

def get_suf_depois(df, cerv, sem, reg_list):
    sub = df[(df["Cerveja"]==cerv) & (df["Semana"]==sem)].set_index("Regiao")
    return [round(float(sub.loc[r,"Suf_f"]),1) if r in sub.index else 0 for r in reg_list]

def get_suf_antes(df, cerv, sem, reg_list):
    sub = df[(df["Cerveja"]==cerv) & (df["Semana"]==sem)].set_index("Regiao")
    vals = []
    for r in reg_list:
        if r not in sub.index:
            vals.append(0); continue
        row     = sub.loc[r]
        ef_orig = pd.to_numeric(row.get("EF_Semana", 0), errors="coerce") or 0
        dem_prox= pd.to_numeric(row.get("Dem_Proxima", row.get("Demanda", 1)), errors="coerce") or 1
        vals.append(round(float(ef_orig) / (float(dem_prox) / 6), 1))
    return vals

antes  = {}
depois = {}
for c in cervejas:
    key = c.strip()
    antes[key]  = {}
    depois[key] = {}
    for s in semanas:
        antes[key][s]  = get_suf_antes(df_all,   c, s, regioes)
        depois[key][s] = get_suf_depois(df_final, c, s, regioes)

cervejas_k = [c.strip() for c in cervejas]

# ── Monta DataFrames longos ───────────────────────────────────────────────────
rows_a, rows_d = [], []
for c in cervejas_k:
    for s in semanas:
        for ri, r in enumerate(regioes):
            rows_a.append({"cerveja": c, "semana": s, "regiao": r, "suf": antes[c][s][ri]})
            rows_d.append({"cerveja": c, "semana": s, "regiao": r, "suf": depois[c][s][ri]})

df_a = pd.DataFrame(rows_a)
df_d = pd.DataFrame(rows_d)

# ── Cores ─────────────────────────────────────────────────────────────────────
COR = {"patagonia": "#378ADD", "malzbier": "#1D9E75", "colorado": "#BA7517", "goose": "#D4537E"}

def suf_color(v):
    if v < 8:   return "#E24B4A"
    if v < 10:  return "#F09595"
    if v < 12:  return "#FAC775"
    if v < 16:  return "#85B7EB"
    return "#1D9E75"

# ═══════════════════════════════════════════════════════════════════════════════
# FIGURA 1 — Histogramas antes vs depois
# ═══════════════════════════════════════════════════════════════════════════════
bins   = [0, 4, 8, 10, 12, 16, 40]
labels = ["0–4", "4–8", "8–10", "10–12", "12–16", "16+"]
cores_hist = ["#E24B4A", "#F09595", "#FAC775", "#FAC775", "#5DCAA5", "#1D9E75"]

def hist_counts(df):
    counts = []
    for i in range(len(bins)-1):
        counts.append(((df["suf"] >= bins[i]) & (df["suf"] < bins[i+1])).sum())
    return counts

fig1 = make_subplots(rows=1, cols=2,
                     subplot_titles=["Antes da redistribuição", "Depois da redistribuição"])

for col, df_h in enumerate([df_a, df_d], start=1):
    counts = hist_counts(df_h)
    fig1.add_trace(go.Bar(
        x=labels, y=counts,
        marker_color=cores_hist,
        text=counts, textposition="outside",
        showlegend=False
    ), row=1, col=col)
    fig1.add_shape(type="line", x0=-0.5, x1=5.5, y0=0, y1=0,
                   line=dict(color="rgba(0,0,0,0)"), row=1, col=col)

fig1.update_layout(
    title_text="Distribuição de suficiência — antes vs depois",
    height=420, template="plotly_white",
    font=dict(family="Arial", size=12),
)
fig1.update_yaxes(title_text="Nº observações")
fig1.update_xaxes(title_text="Dias de suficiência")
fig1.show()

# ═══════════════════════════════════════════════════════════════════════════════
# FIGURA 2 — Barras agrupadas: média por região
# ═══════════════════════════════════════════════════════════════════════════════
avg_a = df_a.groupby("regiao")["suf"].mean().reindex(regioes)
avg_d = df_d.groupby("regiao")["suf"].mean().reindex(regioes)

fig2 = go.Figure()
fig2.add_trace(go.Bar(name="Antes",  x=regioes, y=avg_a.round(1),
                      marker_color="#B4B2A9", text=avg_a.round(1), textposition="outside"))
fig2.add_trace(go.Bar(name="Depois", x=regioes, y=avg_d.round(1),
                      marker_color="#378ADD", text=avg_d.round(1), textposition="outside"))
fig2.add_hline(y=12, line_dash="dash", line_color="#E24B4A",
               annotation_text="meta 12 dias", annotation_position="top right")
fig2.update_layout(
    title="Suficiência média por região — antes vs depois",
    barmode="group", height=420, template="plotly_white",
    yaxis_title="Dias", yaxis_range=[0, max(avg_d.max(), avg_a.max()) * 1.2],
    font=dict(family="Arial", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig2.show()

# ═══════════════════════════════════════════════════════════════════════════════
# FIGURA 3 — Regiões abaixo da meta por semana
# ═══════════════════════════════════════════════════════════════════════════════
below_a = df_a.groupby("semana")["suf"].apply(lambda x: (x<12).sum()).reindex(semanas)
below_d = df_d.groupby("semana")["suf"].apply(lambda x: (x<12).sum()).reindex(semanas)

fig3 = go.Figure()
fig3.add_trace(go.Bar(name="Antes",  x=semanas, y=below_a,
                      marker_color="#F09595", text=below_a, textposition="outside"))
fig3.add_trace(go.Bar(name="Depois", x=semanas, y=below_d,
                      marker_color="#5DCAA5", text=below_d, textposition="outside"))
fig3.update_layout(
    title="Observações abaixo da meta (suf < 12d) por semana",
    barmode="group", height=380, template="plotly_white",
    yaxis_title="Nº região × cerveja", xaxis_title="Semana",
    font=dict(family="Arial", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig3.show()

# ═══════════════════════════════════════════════════════════════════════════════
# FIGURA 4 — Linhas: suf média por cerveja e semana (depois)
# ═══════════════════════════════════════════════════════════════════════════════
fig4 = go.Figure()
for c in cervejas_k:
    avg_sem = [round(np.mean(depois[c][s]), 1) for s in semanas]
    fig4.add_trace(go.Scatter(
        x=semanas, y=avg_sem, name=c,
        mode="lines+markers+text",
        text=avg_sem, textposition="top center",
        line=dict(color=COR[c], width=2.5),
        marker=dict(size=8)
    ))
fig4.add_hline(y=12, line_dash="dash", line_color="#E24B4A",
               annotation_text="meta 12 dias", annotation_position="top right")
fig4.update_layout(
    title="Suficiência média por cerveja e semana — depois da redistribuição",
    height=420, template="plotly_white",
    yaxis_title="Dias", yaxis_range=[0, 30],
    font=dict(family="Arial", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig4.show()

# ═══════════════════════════════════════════════════════════════════════════════
# FIGURA 5 — Heatmap: suf depois por cerveja+região × semana
# ═══════════════════════════════════════════════════════════════════════════════
eixo_y = [f"{c} | {r}" for c in cervejas_k for r in regioes]
z_vals  = [[depois[c][s][ri] for s in semanas] for c in cervejas_k for ri, r in enumerate(regioes)]

fig5 = go.Figure(go.Heatmap(
    z=z_vals,
    x=semanas,
    y=eixo_y,
    colorscale=[
        [0.0,  "#E24B4A"],
        [0.33, "#FAC775"],
        [0.5,  "#FAC775"],
        [0.6,  "#85B7EB"],
        [1.0,  "#1D9E75"],
    ],
    zmid=12,
    zmin=0, zmax=24,
    text=[[f"{v:.1f}d" for v in row] for row in z_vals],
    texttemplate="%{text}",
    colorbar=dict(title="dias", tickvals=[0,6,12,18,24]),
))
fig5.add_vline(x=-0.5, line_width=0)
fig5.update_layout(
    title="Heatmap de suficiência — depois da redistribuição",
    height=600, template="plotly_white",
    xaxis_title="Semana",
    font=dict(family="Arial", size=11),
    margin=dict(l=180)
)
fig5.show()

# ── Salva tudo num HTML único ─────────────────────────────────────────────────
import plotly.io as pio

path = r"C:\Users\ferna\Downloads\AMBEV\dashboard_suf.html"

htmls = []
for i, fig in enumerate([fig1, fig2, fig3, fig4, fig5]):
    include_js = (i == 0)
    htmls.append(fig.to_html(full_html=False, include_plotlyjs=include_js))

with open(path, "w", encoding="utf-8") as f:
    f.write("<html><head><meta charset='utf-8'><style>body{font-family:Arial;padding:2rem;background:#f5f5f3}.chart{background:#fff;border-radius:12px;padding:1rem;margin-bottom:1.5rem;border:0.5px solid #e0e0d8}</style></head><body>")
    f.write("<h1 style='font-size:20px;font-weight:500;margin-bottom:0.5rem'>Suficiencia de estoque - antes e depois da redistribuicao</h1>")
    f.write("<p style='color:#888;font-size:13px;margin-bottom:1.5rem'>Meta: suf >= 12 dias - todas as cervejas - todas as regioes - W0-W3</p>")
    for h in htmls:
        f.write(f"<div class='chart'>{h}</div>")
    f.write("</body></html>")

print(f"Pronto! Abra: {path}")
```
